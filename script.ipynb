```python
offtank = dataset.query("`Selecione sua classe` == 'Off Tank'").copy()
offtank['Digite seu IP'] = pd.to_numeric(offtank['Digite seu IP'], errors='coerce')

offtank.loc[offtank['Digite seu IP'] > 2500, 'Digite seu IP'] = 0
offtank.loc[offtank['Digite seu IP'].isna(), 'Digite seu IP'] = -1
offtank.loc[offtank['Digite seu IP'] <= 0, 'Você tem MOR?'] = 'Não'

offtank = offtank.sort_values(by=["Você tem MOR?", "Digite seu IP"], ascending=[False, False])

offtank
```
